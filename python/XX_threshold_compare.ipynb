## 閾値除去30, 40, 50μVでの、ERP比較、比較波形出力

### 使用脳波データ

### ・XX_threshold_30_data内、threshold_30.datファイル
### ・XX_threshold_40_data内、threshold_40.datファイル 
### ・XX_threshold_50_data内、threshold_50.datファイル

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
files = ["filename1", "filename2", "filename3",]
```

```python
threshold_dataname = ["threshold_30", "threshold_40", "threshold_50"]
```

```python
with open(f'../dataset/XX_threshold_30_data/threshold_30.dat', mode='rb') as f:
    threshold_30_data = pickle.load(f)
with open(f'../dataset/XX_threshold_40_data/threshold_40.dat', mode='rb') as f:
    threshold_40_data = pickle.load(f)
with open(f'../dataset/XX_threshold_50_data/threshold_50.dat', mode='rb') as f:
    threshold_50_data = pickle.load(f)
```

```python
threshold_30_data = threshold_30_data.drop(columns ='HL', axis=1)
threshold_40_data = threshold_40_data.drop(columns ='HL', axis=1)
threshold_50_data = threshold_50_data.drop(columns ='HL', axis=1)
```

```python
threshold_all = pd.DataFrame({'threshold_30':threshold_30_data.sum(axis = 1).tolist(),
                             'threshold_40':threshold_40_data.sum(axis = 1).tolist(),
                             'threshold_50':threshold_50_data.sum(axis = 1).tolist()},
                            index=threshold_30_data['image_stimulus'].tolist())
threshold_all
```

```python
threshold_all.describe()
```

```python
threshold_all.sum()
```

```python
color = ["red", "green", "blue", "purple", "orange", "pink"]
fig = plt.figure(figsize = (120,150))

plt.suptitle('thereshold_all', fontsize=50)

x = threshold_all.index.tolist()
for dataname in range(len(threshold_dataname)):
    plt.subplot(3, 1, dataname+1)
    plt.title(threshold_dataname[dataname], fontsize=70)
    y = threshold_all[threshold_dataname[dataname]]
    plt.plot(x, y, linestyle='None', marker='o', markersize=20, color = color[dataname])
    plt.ylim(-0.5, 2.5)
    plt.xlabel('images',fontsize=50)
    plt.xticks(fontsize=20, rotation=90)
    plt.ylabel('threshold removal',fontsize=50)
    plt.yticks(fontsize=50)

plt.show()
fig.savefig(f'../dataset/threshold_data/threshold_compare_images_1.png')
```

```python
color = ["red", "green", "blue", "purple", "orange", "pink"]
fig = plt.figure(figsize = (120,150))

plt.suptitle('thereshold_all', fontsize=70)

x = threshold_all.index.tolist()
for dataname in range(len(threshold_dataname)):
    plt.subplot(3, 1, dataname+1)
    plt.title(threshold_dataname[dataname], fontsize=50)
    y = threshold_all[threshold_dataname[dataname]]
    plt.bar(x, y,color = color[dataname])
    
    plt.ylim(0, 2.5)
    plt.xlabel('images',fontsize=50)
    plt.xticks(fontsize=20, rotation=90)
    plt.ylabel('threshold removal',fontsize=50)
    plt.yticks(fontsize=50)

plt.show()
fig.savefig(f'../dataset/threshold_data/threshold_compare_images_2.png')
```

```python
threshold_all_sub = pd.DataFrame({'threshold_30':threshold_30_data.sum(axis = 0).tolist()[1:],
                             'threshold_40':threshold_40_data.sum(axis = 0).tolist()[1:],
                             'threshold_50':threshold_50_data.sum(axis = 0).tolist()[1:]},
                            index=files)
threshold_all_sub
```
